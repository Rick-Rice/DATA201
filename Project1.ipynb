## Simulate a nonparametric inference of the Median for a quantitative variable

Multiple samples, leaving replace=True as the default, are taken.  Sample size is about 10% of the entire 2021 data and 50 samples are taken.  The median value of each sample is saved.  The median of the sample medians is 35.0, which -- as luck would have it -- matches the median previously taken of the entire 16,173 rows.  The mean of the sample medians is also 35. 

```python
import numpy as np
 
sample_median = []
for i in range(50):
  y = np.random.choice(crash_reports_2021['Speed Limit'].tolist(),1617)
  #  y = np.random.random.sample(crash_reports_2021['Speed Limit'].tolist(),1617)
  med = np.median(y)
  sample_median.append(med)
 
print("Median of sample medians:",np.median(sample_median))
print("Mean of sample medians:  ",np.mean(sample_median))
```
